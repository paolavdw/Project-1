```python
#import dependencies
import pandas as pd
import matplotlib as plt
import numpy as np
import scipy as sc
from pathlib import Path 

#import dataset from csv file
file = Path(r'C:\Users\lyons\Downloads\PhysicalActivityObesity.csv')
data_df = pd.read_csv(file)

#preview of dataframe
#data_df.head()
#Only need Location, Age and Question columns
data_df = data_df[['LocationAbbr','Age(years)', 'Question']]

#Filter on Question == "Percent of adults who engage in no leisure-time physical activity"
filtered_data_df = data_df.query('Question == "Percent of adults who engage in no leisure-time physical activity"')

#Count total observations by variable 'Age(years)' and 'Location'
group_df = filtered_data_df.groupby(by=['LocationAbbr','Age(years)']).count()
#group_df.head(400)
group_df.to_csv(r'C:\Users\lyons\Downloads\PhysicalActivityObesityGroupBy.csv')
```

```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy as sc
from pathlib import Path 

# reading the database
data = pd.read_csv(r'C:\Users\lyons\Downloads\PhysicalActivityObesityGroupBy10States.csv')

data['Location-Age'] = data['LocationAbbr'].astype(str) + ' ' + data['Age(years)'].astype(str)
#data.head()

counts = data["Question"]
locationage = data["Location-Age"]
# Creating plot
fig = plt.figure(figsize=(10, 7))
plt.pie(counts, labels=locationage)
 
# show plot
plt.show()
```
